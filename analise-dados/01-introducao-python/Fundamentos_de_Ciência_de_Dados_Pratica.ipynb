### 💎 A Metodologia *Orquestra*: Spec-Driven Development (SDD)
Nenhuma linha de código ou modelo preditivo complexo deve ser iniciado sem uma especificação rigorosa (Spec). No futebol moderno, agir de forma reativa aos dados gera análises genéricas. O *Spec-Driven Development* propõe:
1. **Contrato Analítico**: Antes de codificar, especifique detalhadamente qual problema tático ou de mercado estamos resolvendo.
2. **Auditoria Contínua**: Validar a qualidade e o viés dos dados (ex: entender as diferenças metodológicas entre fornecedores como Opta e Wyscout) e garantir o alinhamento com o modelo de jogo da equipe.

### 🧐 Assertividade Socrática
O grande diferencial do Cientista de Dados de elite. Antes de aplicar algoritmos ou criar visualizações, **devemos submeter o problema a um interrogatório socrático**:
* *Por que queremos visualizar os passes progressivos deste atleta da Série B? É porque o modelo de jogo do nosso treinador exige zagueiros que quebram linhas?*
* *Quais as restrições e o contexto do adversário? Ele tem bons números porque joga contra blocos baixos e passivos?*

> **💡 Regra de Ouro**: É melhor fazer 5 perguntas assertivas do que construir 1 modelo de *Machine Learning* inteiro baseado em premissas táticas erradas. **Nunca entregue uma análise sem validar exaustivamente o 'porquê'.**

### 🕸️ Context Intelligence (Graphify)
Um clube de futebol possui silos de dados em *Scouting*, *Fisiologia*, *Tática* e *Mercado*. Um ecossistema AI-First utiliza Grafos de Conhecimento (*Knowledge Graphs*) para cruzar de forma inteligente essas camadas e gerar insights profundos:
* **Exemplo Prático**: Em vez de apenas cruzar métricas básicas, a Engenharia de Contexto permite entender automaticamente o impacto da *carga de acelerações no GPS (Fisiologia)* na *taxa de acerto de cruzamentos (Scouting)* nos *últimos 15 minutos de jogos decisivos (Contexto)*.

# Exemplo Dois: Comparando Mapas de Toque

"A posição da bola do jogador A foi semelhante à do jogador B?"

```python
player1="Gérson"
player2="Carlos Alberto"
touches=['Pass', 'Ball Receipt*', 'Carry', 'Clearance', 'Foul Won', 'Block',
'Ball Recovery', 'Duel', 'Dribble', 'Interception', 'Miscontrol', 'Shot']
```

```python
player1_df=events_df[(events_df.player==player1)&(events_df.type.isin(touches))]
player2_df=events_df[(events_df.player==player2)&(events_df.type.isin(touches))]
```

```python
# Cores do mapa
colour1="white"
colour2="#c3c3c3"
colour3="#e21017"
cmaplist = [colour1, colour2, colour3]
cmap = LinearSegmentedColormap.from_list("", cmaplist)
```

```python
path_eff = [path_effects.Stroke(linewidth=3, foreground='black'),
            path_effects.Normal()]

# Crie uma grade de campos
# Defina o estilo do campo
pitch = VerticalPitch(pitch_type='statsbomb', line_zorder=2, line_color='#000000', linewidth=2, half=False)

# Crie uma grade - neste exemplo, uma linha de dois campos
fig, axs = pitch.grid(nrows=1, ncols=2,
                      figheight=16,
                      grid_width=0.65,
                      endnote_height=0.03, endnote_space=0.05,
                      axis=False, title_space=0.02, title_height=0.06, grid_height=0.8)

# Obtenha dados para o mapa de calor
# jogador1
bin_statistic1 = pitch.bin_statistic(player1_df.x, player1_df.y, statistic='count',
bins=(6, 4), normalize=True)
# jogador2
bin_statistic2 = pitch.bin_statistic(player2_df.x, player2_df.y, statistic='count',
bins=(6, 4), normalize=True)

# Calcule o valor vmax
# queremos que ambos os nossos mapas de calor usem o mesmo espectro de cores
vmax1 = bin_statistic1['statistic'].max()
vmax2 = bin_statistic2['statistic'].max()

if vmax1 > vmax2:
    vmax = vmax1
else:
    vmax = vmax2

# Defina vmin como zero
vmin = 0

# Plote o mapa de calor para o jogador1
heatmap1 = pitch.heatmap(bin_statistic1, ax=axs['pitch'][0], cmap=cmap, vmax=vmax, vmin=vmin)

# Adicione porcentagens para cada zona
annotate1 = pitch.label_heatmap(bin_statistic1, color='white',
                                path_effects=path_eff, fontsize=50, ax=axs['pitch'][0],
                                str_format='{:.0%}',
                                ha='center', va='center',
                                exclude_zeros=True)
# Abrevuação nomes
player1 = 'Gérson'
player2 = 'Carlos Alberto'

# Adicione o nome do jogador ao topo do mapa de calor
ax_text(0, 125, f'{player1}: Toques', ha='left', c='black', va='center', fontsize=45, ax=axs['pitch'][0], style='italic')


# Plote o mapa de calor para o jogador2
heatmap2 = pitch.heatmap(bin_statistic2, ax=axs['pitch'][1], cmap=cmap, vmax=vmax, vmin=vmin)

# Adicione porcentagens para cada zona
annotate2 = pitch.label_heatmap(bin_statistic2, color='white',
                                path_effects=path_eff, fontsize=50, ax=axs['pitch'][1],
                                str_format='{:.0%}',
                                ha='center', va='center',
                                exclude_zeros=True)

# Adicione o nome do jogador ao topo do mapa de calor
ax_text(0, 125, f'{player2}: Toques', ha='left', c='black', va='center', fontsize=45, ax=axs['pitch'][1], style='italic')
```

# Bônus

Obtenha os melhores desempenhos de xG ao longo do torneio. Crie um mapa de chutes para o melhor jogador.

```python
# Obtenha os dados de eventos para toda a temporada
comp_events_df = sb.competition_events(
     country='Europe',
     division='UEFA Euro',
     season='2024',
     gender="male")
```

```python
comp_events_df[['x', 'y']] = comp_events_df['location'].apply(pd.Series)
comp_events_df[['pass_end_x', 'pass_end_y']] = comp_events_df['pass_end_location'].apply(pd.Series)
comp_events_df[['carry_end_x', 'carry_end_y']] = comp_events_df['carry_end_location'].apply(pd.Series)
```

```python
# Obtenha chutes que não são pênaltis e gols que não são pênaltis
shots_df = comp_events_df[(comp_events_df.type == "Shot") & (comp_events_df.shot_type != "Penalty")]
goals_df = comp_events_df[(comp_events_df.type == "Shot") & (comp_events_df.shot_type != "Penalty") & (comp_events_df.shot_outcome == "Goal")]

# Chutes
# Obtenha a contagem para cada jogador, incluindo informações do time
shots_count = shots_df.groupby(['player', 'team']).size().reset_index()

# Renomeie a segunda coluna
shots_count.rename(columns={shots_count.columns[2]: "Chutes np" }, inplace=True)

# Gols
# Obtenha a contagem para cada jogador, incluindo informações do time
goals_count = goals_df.groupby(['player', 'team']).size().reset_index()

# Renomeie a segunda coluna
goals_count.rename(columns={goals_count.columns[2]: "np Shots" }, inplace=True)

# xG
# Obtenha a soma para cada jogador, incluindo informações do time
xg_sum = shots_df.groupby(['player', 'team'])['shot_statsbomb_xg'].sum().reset_index()

# Renomeie a segunda coluna
xg_sum.rename(columns={xg_sum.columns[2]: "np xG" }, inplace=True)

# Arredonde para 2 casas decimais
xg_sum['np xG'] = xg_sum['np xG'].round(2)
```

```python
#combine dataframes
shot_stats=pd.merge(shots_count, goals_count,
                    how="outer", on=["player", "team"])

shot_stats=pd.merge(shot_stats, xg_sum,
                    how="outer", on=["player", "team"])
```

```python
#add 0 where value is NaN
shot_stats=shot_stats.fillna(0)

#sort to get top goalscorers, change by=' ' to get top xG, or top Shots
shot_stats=shot_stats.sort_values(by='np xG', ascending=False)
```

```python
shot_stats.head(10)
```

```python
# Mapa de chutes para o artilheiro
xg_player=shot_stats.player.iloc[0]

player_goals_df=goals_df[(goals_df.player==xg_player)]

player_shots_df=shots_df[(shots_df.player==xg_player)&(shots_df.shot_outcome!="Goal")]
```

```python
pitch = VerticalPitch(pitch_type='statsbomb',half=True,pad_bottom=-11)
fig, ax = pitch.draw(figsize=(12, 10))

shot_colour='black'
goal_colour='green'

shots = pitch.scatter(player_shots_df.x, player_shots_df.y,
# Torne o tamanho relativo ao valor de xG
s=player_shots_df.shot_statsbomb_xg*1000,
c=shot_colour,
label='Shots',
marker='o',
alpha=0.5,
ax=ax)

goals = pitch.scatter(player_goals_df.x, player_goals_df.y,
# Torne o tamanho relativo ao valor de xG
s=player_goals_df.shot_statsbomb_xg*1000,
c='white',
edgecolors=goal_colour,
label='Goals',
marker='football',
ax=ax)

# Plotar a legenda
ax.legend(facecolor='white', handlelength=5, edgecolor='None', fontsize=20, loc='best')

ax_text(0, 125, f'{xg_player}: Shot Map, Euro 2024', ha='left', c='black',va='center', fontsize=25,ax=ax,style='italic')
```

```python
# Otimizar os nomes
brazil_df['player'] = brazil_df['player'].replace({
            'Édson Arantes do Nascimento': 'Pelé',
            'Carlos Alberto Torres': 'Carlos Alberto',
            'Jair Ventura Filho': 'Jairzinho',
            'Clodoaldo Tavares de Santana': 'Clodoaldo',
            'José Ely de Miranda': 'Zito',
            'Waldyr Pereira': 'Didi',
            'Roberto Rivelino': 'Rivelino',
            'Gérson de Oliveira Nunes': 'Gérson',
            'Mário Jorge Lobo Zagallo': 'Zagallo',
            'Everaldo Marques da Silva': 'Everaldo',
            'Nilton dos Reis Santos': 'Nilton Santos',
            'Dejalma Pereira Dias dos Santos': 'Djalma Santos',
            'Eduardo Gonçalves de Andrade': 'Tostão',
            'Manoel Francisco dos Santos': 'Garrincha',
            'Edvaldo Izidio Neto': 'Vavá',
            'Paulo Cézar Lima': 'Paulo Cézar',
            'Nilton de Sordi': 'De Sordi',
            'Hercules Brito Ruas': 'Brito',
            'Roberto Lopes de Miranda': 'Roberto Miranda',
            'Orlando Peçanha de Carvalho': 'Orlando',
            'Marco Antônio Feliciano': 'Marco Antônio',
            'Mauro Ramos de Oliveira': 'Mauro',
            'Wilson da Silva Piazza': 'Piazza',
            'Gylmar dos Santos Neves': 'Gylmar',
            'Hideraldo Luis Bellini': 'Bellini',
            'Zózimo Alves Calazães': 'Zózimo'
})
```

```python
# Separe as localizações de início e fim das coordenadas
brazil_df[['x', 'y']] = brazil_df['location'].apply(pd.Series)
brazil_df[['pass_end_x', 'pass_end_y']] = brazil_df['pass_end_location'].apply(pd.Series)
brazil_df[['carry_end_x', 'carry_end_y']] = brazil_df['carry_end_location'].apply(pd.Series)
```

```python
# Filtre para passes completados
f3rd_passes = brazil_df[(brazil_df.team == squad) & (brazil_df.type == "Pass") & (brazil_df.x < 80) & (brazil_df.pass_end_x > 80) & (brazil_df.pass_outcome.isna())]

# Obtenha a contagem para cada jogador
f3rd_passes_count = f3rd_passes.groupby('player').size().reset_index()
```

```python
f3rd_passes_count
```

```python
# Renomear a coluna passes
f3rd_passes_count.rename(columns={f3rd_passes_count.columns[1]: "Passes" }, inplace = True)
```

```python
f3rd_passes_count
```

```python
# Obtenha as conduções no terceiro final para o time de foco
f3rd_carries = brazil_df[(brazil_df.type == "Carry") & (brazil_df.x < 80) & (brazil_df.carry_end_x > 80) & (brazil_df.team == squad)]

# Obtenha a contagem para cada jogador
f3rd_carries_count = f3rd_carries.groupby('player').size().reset_index()

# Renomeie a segunda coluna
f3rd_carries_count.rename(columns={f3rd_carries_count.columns[1]: "Conduções"}, inplace=True)
```

```python
f3rd_carries_count
```

```python
#combine dataframes
progressions_df=pd.merge(f3rd_passes_count, f3rd_carries_count,
                    how="outer", on=["player"])
```

```python
#add 0 where value is NaN
progressions_df=progressions_df.fillna(0)
```

```python
# Crie um gráfico de barras empilhadas
pass_colour = '#e21017'
carry_colour = 'blue'

# Ordene para obter os menores valores primeiro (para que seja plotado na ordem correta em nosso gráfico de barras)
progressions_df.sort_values(by='total', ascending=True, inplace=True)

# Inclua apenas as colunas relevantes
barchart_df = progressions_df[["player", "Passes", "Conduções"]]

# Crie a figura e defina alguns parâmetros de estilo
plt.figure(figsize=(15, 15))
sns.set(rc={'axes.facecolor': 'white', 'figure.facecolor': 'white'})
sns.set_style("ticks")

# Crie o gráfico de barras
barchart_df.set_index('player').plot(kind='barh', stacked=True, color=[pass_colour, carry_colour], legend=True, figsize=(10, 10))

# Adicione rótulos e título ao gráfico
plt.xlabel(xlabel="Avanços para o Terceiro Final", fontdict={'fontsize': 12, 'weight': 'semibold'})
plt.ylabel(ylabel="Jogador", fontdict={'fontsize': 12, 'weight': 'semibold'})
plt.title(f"{squad}: Avanços para o Terceiro Final", fontdict={'fontsize': 24, 'weight': 'bold'})

# Remova os eixos superiores e direitos
sns.despine(top=True, right=True, left=False, bottom=False)
plt.autoscale()
plt.show()
```

```python
player1="Diego Armando Maradona"
player2="Jorge Luis Burruchaga"
touches=['Pass', 'Ball Receipt*', 'Carry', 'Clearance', 'Foul Won', 'Block',
'Ball Recovery', 'Duel', 'Dribble', 'Interception', 'Miscontrol', 'Shot']
```

```python
player1_df=brazil_df[(brazil_df.player==player1)&(brazil_df.type.isin(touches))]
player2_df=brazil_df[(brazil_df.player==player2)&(brazil_df.type.isin(touches))]
```

```python
# Cores do mapa
colour1="white"
colour2="#c3c3c3"
colour3="#e21017"
cmaplist = [colour1, colour2, colour3]
cmap = LinearSegmentedColormap.from_list("", cmaplist)
```

```python
path_eff = [path_effects.Stroke(linewidth=3, foreground='black'),
            path_effects.Normal()]

# Crie uma grade de campos
# Defina o estilo do campo
pitch = VerticalPitch(pitch_type='statsbomb', line_zorder=2, line_color='#000000', linewidth=2, half=False)

# Crie uma grade - neste exemplo, uma linha de dois campos
fig, axs = pitch.grid(nrows=1, ncols=2,
                      figheight=16,
                      grid_width=0.65,
                      endnote_height=0.03, endnote_space=0.05,
                      axis=False, title_space=0.02, title_height=0.06, grid_height=0.8)

# Obtenha dados para o mapa de calor
# jogador1
bin_statistic1 = pitch.bin_statistic(player1_df.x, player1_df.y, statistic='count',
bins=(6, 4), normalize=True)
# jogador2
bin_statistic2 = pitch.bin_statistic(player2_df.x, player2_df.y, statistic='count',
bins=(6, 4), normalize=True)

# Calcule o valor vmax
# queremos que ambos os nossos mapas de calor usem o mesmo espectro de cores
vmax1 = bin_statistic1['statistic'].max()
vmax2 = bin_statistic2['statistic'].max()

if vmax1 > vmax2:
    vmax = vmax1
else:
    vmax = vmax2

# Defina vmin como zero
vmin = 0

# Plote o mapa de calor para o jogador1
heatmap1 = pitch.heatmap(bin_statistic1, ax=axs['pitch'][0], cmap=cmap, vmax=vmax, vmin=vmin)

# Adicione porcentagens para cada zona
annotate1 = pitch.label_heatmap(bin_statistic1, color='white',
                                path_effects=path_eff, fontsize=50, ax=axs['pitch'][0],
                                str_format='{:.0%}',
                                ha='center', va='center',
                                exclude_zeros=True)


# Adicione o nome do jogador ao topo do mapa de calor
ax_text(0, 125, f'{player1}: Toques', ha='left', c='black', va='center', fontsize=45, ax=axs['pitch'][0], style='italic')


# Plote o mapa de calor para o jogador2
heatmap2 = pitch.heatmap(bin_statistic2, ax=axs['pitch'][1], cmap=cmap, vmax=vmax, vmin=vmin)

# Adicione porcentagens para cada zona
annotate2 = pitch.label_heatmap(bin_statistic2, color='white',
                                path_effects=path_eff, fontsize=50, ax=axs['pitch'][1],
                                str_format='{:.0%}',
                                ha='center', va='center',
                                exclude_zeros=True)

# Adicione o nome do jogador ao topo do mapa de calor
ax_text(0, 125, f'{player2}: Toques', ha='left', c='black', va='center', fontsize=45, ax=axs['pitch'][1], style='italic')
```
