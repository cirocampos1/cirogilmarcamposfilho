### Operadores Matemáticos
Como calculamos métricas simples? Usamos `+`, `-`, `*`, `/`.

## 📊 Módulo 2: Estruturas de Dados e Fluxo

No mundo real, não temos apenas "1 jogador". Temos escalações inteiras!
Para lidar com vários dados, usamos **Listas** e **Dicionários**.

```python
escalacao = ["Alisson", "Danilo", "Marquinhos", "Militão", "Wendell", "Casemiro", "Bruno Guimarães", "Paquetá", "Raphinha", "Vini Jr", "Rodrygo"]

print(f"Goleiro: {escalacao[0]}") # Em Python, contamos a partir do ZERO!
print(f"Ataque: {escalacao[8:]}") # Pegando do 8º jogador em diante
```

### Fluxo Condicional (If / Elif / Else)
Como programamos o computador para tomar decisões analíticas?
Exemplo: Um scout precisa avaliar se um jogador deve ser contratado baseado na sua nota (Rating).\n

```python
nota_sofascore = 7.8

if nota_sofascore >= 8.0:
    print("Contratar imediatamente! (Classe A)")
elif nota_sofascore >= 7.0:
    print("Monitorar desempenho. (Classe B)")
else:
    print("Abaixo da média da liga. (Classe C)")
```

### Laços de Repetição (`for`)
Analisar dados muitas vezes significa olhar para muitos dados. O `for` repete uma ação para cada item em uma lista.\n

```python
# Analisando a quantidade de passes de vários meio-campistas
passes_por_jogador = [45, 60, 32, 88, 12]

total_passes_time = 0
for passes in passes_por_jogador:
    total_passes_time += passes
    if passes > 50:
        print(f"Motorzinho do time! Fez {passes} passes.")
        
print(f"Total de passes da equipe: {total_passes_time}")
```

## ⚙️ Módulo 3: Funções (Modularização)

Se você repete o mesmo cálculo de performance o tempo todo, crie uma **Função**.
Uma função é uma fábrica: você entra com matéria-prima (Parâmetros) e sai com um produto final (Retorno).\n

```python
def calcular_minutos_por_gol(minutos_jogados, gols):
    # Tratamento de erro básico (Não dividir por zero)
    if gols == 0:
        return "Não marcou"
    
    return round(minutos_jogados / gols, 1)

# Testando a função com dois atacantes
minutos_a1 = calcular_minutos_por_gol(450, 5) # 5 gols em 450 min
minutos_a2 = calcular_minutos_por_gol(600, 0) # 0 gols em 600 min

print(f"Atacante 1 precisa de {minutos_a1} mins para fazer um gol.")
print(f"Atacante 2: {minutos_a2}")
```

### Tratamento de Erros Moderno (`try / except`)
Às vezes recebemos dados sujos da API e o Python quebra (Erro). O `try / except` nos protege.\n

```python
api_data = {"passes": 50, "certos": "Quarenta"} # O dado veio como texto!

try:
    precisao = api_data['certos'] / api_data['passes']
    print(precisao)
except TypeError:
    print("ERRO DE DADOS: O número de passes certos veio como texto (string). Precisamos higienizar a base!")
except KeyError:
    print("ERRO DE DADOS: A chave 'certos' não existe no JSON.")
```

## 🐼 Módulo 4: Ponte para Ciência de Dados (Introdução ao Pandas)

Você chegou no nível intermediário! Na vida real da análise de dados de futebol, não iteramos listas com o comando `for` toda hora. Nós usamos a biblioteca **Pandas**. Ela é como um Excel com "esteróides".

O formato tabular principal do Pandas se chama **DataFrame**.\n

```python
# O Pandas não vem nativo, precisamos importá-lo!
import pandas as pd
from IPython.display import display

# Simulando um extrato tático extraído do Sofascore
dados = {
    "Jogador": ["Ederson", "Casemiro", "Paquetá", "Vini Jr", "Endrick"],
    "Posicao": ["GK", "VOL", "MEI", "ATA", "ATA"],
    "Nota": [7.1, 8.2, 7.5, 9.0, 8.0],
    "Gols": [0, 0, 1, 2, 1],
    "Passes_Tentados": [25, 80, 60, 20, 10]
}

# Transformando o Dicionário em um DataFrame do Pandas
df = pd.DataFrame(dados)

# .head() exibe as 5 primeiras linhas da "planilha"
display(df.head())
```

### Filtros Mágicos do Pandas
Filtramos dezenas de milhares de linhas instantaneamente.\n

```python
# Filtrar apenas os Atacantes
atacantes = df[df['Posicao'] == 'ATA']
display(atacantes)
```

```python
# Filtrar jogadores com Nota maior ou igual a 8.0
craques = df[df['Nota'] >= 8.0]
display(craques)
```

```python
# Criando uma coluna nova calculada! (Ex: Participações Diretas = Gols + 0 Assistencias pro exemplo)
df['Participacoes'] = df['Gols'] * 1 
display(df)
```

### Agrupamentos Simples
Quem é a média de notas por Posição? (Groupby)\n

```python
media_por_posicao = df.groupby('Posicao')['Nota'].mean()
display(media_por_posicao)
```
